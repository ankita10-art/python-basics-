2. Explain the difference between List and Tuple in Python?
- In python, both lists and tuples are used to store multiple items in a single variable . however, they have several important differences based on their properties and usage.

# Defination
 A List is a collection of elements that is mutable , meaning its value can be changed after creation .
 A tuple is a collection of elements that is immutable , meaning its values cannot be changed once defined.

# syntax
Lists are created using square brackets []
whereas tuples are created using paretheses()
Example:
. List - [1,2,3]
. Tuple- (1,2,3)

# Mutability
Lists are mutable ,so elements can be added , removed , or modify .
tuples are immitable , so elements remain constant throghout the program .

# Performance
 tuples are geneeally faster than lists because they are fixed and require less memory . Lists are slightly slower due to their dynamic nature .

# Methods
Lists have many buil-in methods such as append (),remove (),and sort().
tuples have limited methods like count() and index().

# Use cases
Lists are used when data needs to be modified frequently .
Tuple are used when data should remain constant and secure .

lists and tuples area both useful data structures in python . Lists provide flexibility due to their mutable nature , while tuples ensure data integrity because they cannot be changed . the choice between them depends on the requirement of the program .

8. Write a Python program to remove duplicate values from a list using a set?
- AIM:
To write a python program to remove duplicate values from a list using a set .


```python
#create a list with duplicate values
numbers = [10,20,10,20,40,30,40]

#remove duplicate using set
unique_numbers = list(set(numbers))

#print result
print("list after removing duplicates:",unique_numbers)
```

explanation:
A list numbers contains duplicaaate values.
set() automatically removes duplicates elements.
list() converts the set back into a list .
print () displays the final list.

Output:
List after removing duplicates : [40,10,20,30]
